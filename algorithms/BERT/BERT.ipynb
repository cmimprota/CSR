# 1 Data Preprocessing

### 1.1 Load data

```python
f = open("/home/xiaochenzheng/Desktop/cil-spring20-project-data_preprocessing/twitter-datasets/train_neg.txt")
tweets = f.readlines()
f.close()
```

### 1.2 Remove HTML tag

```python
htmltag = re.compile('<.*?>')
for i, _tweet in enumerate(tweets):
    _tweet = re.sub(htmltag, ' ', _tweet) # remove <user><url>
    tweets[i] = _tweet
```

### 1.3 Remove Punctuation, Digits, and Keep Hashtag

```python
for i, _tweet in enumerate(tweets):
    _tweet = re.sub('[^a-z A-Z0-9_#]',' ',_tweet) # Keep hashtag, remove punctuation
    _tweet = re.sub(' \d+',' ',_tweet) # Remove digits
    tweets[i] = _tweet
```

### 1.4 Remove Stopwords and Optional: Remove word length smaller than two

```python
# Incorrect stopwords
incorrect_stopwords = ['im', "shes", 'whats', 'thats', "dont", "arent", "couldnt", "didnt", "doesnt", "hadnt", "hasnt", "havent", "isnt", "mightnt", "mustnt", "neednt", "shant", "shouldnt", "wasnt", "werent", "wont", "wouldnt", 'cannot']
```

```python
for i, _tweet in enumerate(tweets):
    tidy = [w for w in _tweet.split() if w not in stopwords.words('english') and w not in incorrect_stopwords and len(w)>=3]
    tweets[i] = tidy
```

### 1.5 Lemmatization

* SnowballStemmer: leaves ---> leav
* WorldNetLemmatizer: leaves ---> leaf

```python
from nltk.stem import WordNetLemmatizer 
```

```python
wnl = WordNetLemmatizer() 
for i, _tweet in enumerate(tweets):
    tidy = [wnl.lemmatize(w) for w in _tweet]
    tweets[i] = tidy
```
