```python
import wordsegment
import pandas as pd
import re
import numpy as np
```

```python
# LOAD DATA

f = open("./twitter-datasets/train_pos.txt")
tweets = f.readlines()
f.close()
```

```python
tags = [];
for tweet in tweets:
    s = {tag.strip("#") for tag in tweet.split() if tag.startswith("#")}
    
    tags.append(list(s))
tags = list(filter(None, tags))
```

```python
tags
```

```python
wordsegment.load()
```

```python
clean_tags = []

for tweets in tags:
    for tag in tweets:
        clean_tags.append(wordsegment.segment(tag))
```
